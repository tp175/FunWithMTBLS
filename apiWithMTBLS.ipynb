# MetaboLights
https://www.ebi.ac.uk/metabolights

```python
#xxx
#https://www.ebi.ac.uk/metabolights/ws/api/spec.html#!/spec

#url = "https://www.ebi.ac.uk/metabolights/ws/studies"
url = "https://www.ebi.ac.uk/metabolights/ws/studies/technology"
print(url)

response = requests.get(url)
#print(response)
rjson = response.json()
#rjson = response.content
print(rjson)
```

```python
#datClip = response.json()['content']
#df = pd.DataFrame(datClip, columns=['study_id'])

datClip = rjson
df = pd.DataFrame(datClip)
df.columns = ['study_id','study_type']

df.to_excel('datMTBLSstudy_'+cDATE+'.xlsx', sheet_name='Sheet1', index=True)
print(df.head(n=5))
print(df.tail(n=6))
print(df.shape)
```

```python
# skip (limit to test)
n = 5
df = df[0:n]
print(df)
```

```python
#from itertools import product
#df = df[0:5]
#df = df[-6:]

study_data_labels = 'derivedData'
#study_data_labels = ['derivedData','organism']

Study_ID = df['study_id']
study_title = []
study_data = []
study = []

for i in range(0, len(Study_ID)):
  print(i)
  url = "https://www.ebi.ac.uk/metabolights/ws/studies/public/study/" + Study_ID[i]
  #https://www.ebi.ac.uk/metabolights/ws/studies/public/study/MTBLS3
  #print(url)

  response = requests.get(url)
  #print(response)
  rjson = response.json()
  #rjson = response.content
  #print(rjson)

  study_title.append(rjson['content']['title'])
  study_data.append(rjson['content'][study_data_labels])
  study.append(rjson['content'])
  #Organism_Part_Pair.append([Study_ID[i],list(product(Species_Organism,Source_Organism_Part))])


df['study_title'] = study_title
df2 = pd.concat([df, pd.DataFrame(study_data)], axis=1)
df2.to_excel('datMTBLSstudy_'+cDATE+'.xlsx', sheet_name='Sheet1', index=True)
print(df2.head(n=3))
print(df2.shape)
```

# Metabolomics Workbench
https://www.metabolomicsworkbench.org

```python
#https://www.metabolomicsworkbench.org/tools/MWRestAPIv1.0.pdf
#https://www.metabolomicsworkbench.org/tools/mw_rest.php

url = "https://www.metabolomicsworkbench.org/rest/study/study_id/ST/available"
#url = "https://www.metabolomicsworkbench.org/rest/study/study_id/ST/summary" #...TIMES OUT...
#url = "https://www.metabolomicsworkbench.org/rest/study/study_id/X/untarg_studies"
print(url)

response = requests.get(url)
#print(response)
rjson = response.json()
#rjson = response.content
print(rjson)
```

```python
datClip = rjson
df = pd.DataFrame(datClip).transpose()
df.to_excel('datMWstudy_'+cDATE+'.xlsx', sheet_name='Sheet1', index=True)

print(df.head(n=5))
print(df.tail(n=6))
print(df.shape)
```

```python
#from itertools import product
#df = df[0:5]
#df = df[-6:]

#Study_ID = df['study_id']
Study_ID = df['study_id'].unique()
study_title = []
study_id = []
study = []

for i in range(0, len(Study_ID)):
  print(i)
  url = "https://www.metabolomicsworkbench.org/rest/study/study_id/" + Study_ID[i] + "/summary"
  #https://www.metabolomicsworkbench.org/rest/study/study_id/ST000001/summary
  #print(url)

  response = requests.get(url)
  #print(response)
  rjson = response.json()
  #rjson = response.content
  #print(rjson)

  try:
    # old?
    #print(rjson['study_title'])
    study_title.append(rjson['study_title'])
    study_id.append(rjson['study_id'])
    study.append(rjson)
  except:
    # new?
    #print(rjson['1']['study_title'])
    study_title.append(rjson['1']['study_title'])
    study_id.append(rjson['1']['study_id'])
    study.append(rjson['1'])
  #Organism_Part_Pair.append([Study_ID[i],list(product(Species_Organism,Source_Organism_Part))])


#dfU = pd.DataFrame([study_id,study_title,study_type]).transpose()
#dfU.columns = ['study_id','study_title','study_type']
dfU = pd.DataFrame(study).transpose()
dfU.to_excel('datMWstudy_'+cDATE+'.xlsx', sheet_name='Sheet1', index=True)
print(dfU.head(n=3))
print(dfU.shape)
```
